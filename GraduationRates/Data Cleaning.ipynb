```python
# Selecting categorical columns
categorical_df = df.filter(regex="Carnegie|Historically|Sector|size category|(FLAGS2019)").copy(deep=True)
```

```python
categorical_df.head()
```

```python
# Using Groupby object from schema to fill categorical variables with their string values
for i in categorical_df.columns:
    val_label = groups.get_group(i)[["Value", "ValueLabel"]].transpose()

    val_label.columns = val_label.loc["Value"]

    val_label.drop("Value", inplace=True)

    map_ = val_label.loc["ValueLabel"].to_dict()

    map_ ={int(i): v for i,v in map_.items()}
    map_[-100] = "isMissing"
    
    categorical_df[i] = categorical_df[i].fillna(-100).astype(int).replace(map_)
```
